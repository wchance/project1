```python
# imports
import requests
import json
import panel as pn
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
from dotenv import load_dotenv

import warnings
warnings.filterwarnings('ignore')
```

```python
dow_jones_list = pd.read_json ('../data/rawdata/dow_jones_list.json')
dow_jones_list.head(3)
```

```python
# Sunburst Plot Sector Distributions Dow Jones
dj_sector_distribution_plot = px.sunburst(
     dow_jones_list,
     path=['sector'],
     width=600,height=600,
    title='Dow_Jones Sector % Distribution',
)
dj_sector_distribution_plot.show()
```

```python
# Sunburst Plot Headquarter Distributions Dow Jones
dj_hq_distribution_plot = px.sunburst(
     dow_jones_list,
     path=['headQuarter'],
     width=600,height=600,
    title='Dow_Jones Headquarter Distribution',
)
dj_hq_distribution_plot.show()
```

```python
dow_jones_sectors=dow_jones_list.groupby('sector').count()
dow_jones_sectors.reset_index(drop=False,inplace=True)
dow_jones_sectors
```
